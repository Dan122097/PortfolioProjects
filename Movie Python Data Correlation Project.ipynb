```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Read in the data

df = pd.read_csv(r'C:\Users\dieha\Documents\JupyterNotebookProjects\MovieCorrelation\movies.csv')

```

```python
# Checking out the data
df.head()
```

```python
# Loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
# Remove missing values
df = df.dropna()
```

```python
# Confirming missing data removed

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data types of columns to remove unneccesary decimal points

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Year column seems to differ from year in released in column
# Create correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
# Order the data descending by gross income

df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
# Checking for any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Let's check for any outliers

df.boxplot(column=['gross'])
```

```python
# Testing out the hypothesis:
# If a movie has a higher budget, then the gross income will become higher as well
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')


plt.show()
```

```python
# Adding in a linear regression line
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Correlation Matrix 

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
```

```python
# Correlation Matrix with features quantified

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
